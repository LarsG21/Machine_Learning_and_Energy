- $f_1(\mathbf{x}) = \exp\left(\frac{(x_1-2)^2}{5}+\frac{(x_2-3)^2}{10}-1\right)$,
- $f_2(\mathbf{x}) = \sin(\pi(x_1-1))+\sin(\pi(x_2-1))$,

- $f_4(\mathbf{x}) = h(\mathbf{x},(3\; 2)^T,-\pi/8,50)$,

$f_3(\mathbf{x})$ and $f_4(\mathbf{x})$ are defined by the function $h$:

$h(\mathbf{x},\mathbf{c},\alpha,\lambda) = (\mathbf{x}-\mathbf{c})^T\mathbf{A}_{\alpha,\lambda}(\mathbf{x}-\mathbf{c})$ with $\mathbf{A}_{\alpha,\lambda} = \mathbf{R}_\alpha^T\mathbf{diag}(1 \;\lambda)\mathbf{R}_\alpha$ and $\mathbf{R}_\alpha = \begin{pmatrix} \cos(\alpha) & -\sin(\alpha)\\ \sin(\alpha) & \cos(\alpha)\end{pmatrix}$.

In each iteration the algorithm should compute the function value at the point $x_0+t$ and double the step size $t$ as long as the new function value is still smaller than the old one.<br>
Choose $t=10^{-4}$ as initial step size. Use ``while True:`` to create a loop and an ``if``-condition plus ``break`` to stop the loop once the values stop decreasing or it performed more than ``maxiter=25`` iterations.

#### b) Run the following cell to test your implementation on the function $f(z)=(z-1)^2$ with starting point $z_0=-5$.

```python
from linesearch import linesearch
import numpy as np

z = np.arange(-10, 10, step=0.1)
z_0 = -5
def f_test(z): return (z-1)**2

t = linesearch(f_test, z_0) # get step size from linesearch function

plt.figure(figsize=[15,7])
plt.plot(z, f_test(z), color='black', label='$f_{test}$')
plt.plot(z_0, f_test(z_0), 'bo', label='start point')
plt.plot(z_0+t, f_test(z_0+t), 'ro', label='end point')
plt.plot(1, 0, 'kx', label='minimum')
plt.legend()
_ = plt.title('Line search')
```

## 4. Using built-in optimizers
Although implementing a simple descending algorithm is useful for understanding purposes, in practice we prefer to use in-built functions. The reason for that is two-fold: the algorithm's implementation has already been tested by the whole community. Thus, it is less error-prone than self-implemented ones; libraries such as scipy provide more powerful algorithms with much faster convergence.<br>
#### a) Complete code in the cell below to optimize our functions using ``minimize()`` from ``scipy.optimize``. Store the found optimums in ``y_opt_scipy``.
Loop over the same ``ids`` and the same starting points ``x_0`` used in the previous cell.

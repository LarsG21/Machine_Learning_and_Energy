Paper abstract:

*Clouds have an enormous influence on the Earth's energy balance, climate, and weather. Cloud types have different cloud radiative effects, which is an essential indicator of the cloud effect on radiation. Therefore, identifying the cloud type is important in meteorology. In this letter, we propose a new convolutional neural network model, called CloudNet, for accurate ground‐based meteorological cloud classification. We build a ground‐based cloud data set, called Cirrus Cumulus Stratus Nimbus, which consists of 11 categories under meteorological standards. The total number of cloud images is three times that of the previous database. In particular, it is the first time that contrails, a type of cloud generated by human activity, have been taken into account in the ground‐based cloud classification, making the Cirrus Cumulus Stratus Nimbus data set more discriminative and comprehensive than existing ground‐based cloud databases. The evaluation of a large number of experiments demonstrates that the proposed CloudNet model could achieve good performance in meteorological cloud classification.*

```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
tf.keras.backend.clear_session()
tf.random.set_seed(123)
```

## 1. Load the data set
*NOTE: The cloud image data set is available via a separate download link from Moodle. Download the data set, unzip the file, and place the folder in the directory of this notebook or your working directory.*
#### a) Run the cell to load the data.

```python
N_IMAGES = 2543
class_names = ["Ac","As","Cb","Cc","Ci","Cs","Ct","Cu","Ns","Sc","St"]
image_size = (64,64)

train_ds=tf.keras.preprocessing.image_dataset_from_directory(
    "CCSN_v2",
    labels="inferred",
    label_mode="categorical",
    class_names=class_names,
    color_mode="rgb",
    batch_size=N_IMAGES,
    image_size=image_size,
    shuffle=True,
    seed=123)
```

#### b) Run the cell below to plot a random image from each class.

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[np.nonzero(labels.numpy()[i])[0][0]])
        plt.axis("off")
```

#### c) Run the cell below to split the data in training and test set. All values will be scaled to lie between 0 and 1.

```python
for X,y in train_ds.as_numpy_iterator():
    pass

X_train = X[0:2300]/255.0
y_train = y[0:2300]

X_test = X[2300:]/255.0
y_test = y[2300:]
```

d) Have look at the values in ``y``. How are the labeles encoded?

## 2. Model training
#### a) Complete the code in the function ``build_model()`` in the module ``CloudNet.py``.
Implement the following model:
- Convolutional Layer: 16 filters, 8x8 kernel, relu activation
- Max Pooling: 4x4
- Convolutional Layer: 32 filters, 4x4 kernel, relu activation
- Max Pooling: 2x2
- Convolutional Layer: 64 filters, 4x4 kernel, relu activation
- Max Pooling: 2x2
- Dense layer: 11 neurons (number of classes), softmax activation <-- this is the output layer

HINT: You will need to "flatten" the output of the pooling layer before the output layer using a flatten layer because the dense layer expects a vevtor and not a matrix.

#### b) Run the cell below to build the model and plot a summary.

```python
from CloudNet import build_model
model = build_model()
model.summary()
img=keras.utils.plot_model(model, "CloudNet.png", show_shapes=True, dpi=96, show_layer_names=True)
img
```

c) In the output layer of the model, we used a softmax function
$$g(z)_i = \frac{exp(z_i)}{\sum_{j=1}^Kexp(z_j)}$$
instead of the usual sigmoid function. Furthermore we used categorial crossentropy given by
$$L = -\sum_{j=1}^K y_jlog(\hat{y_j})$$
as our loss function.
#### What is the reason for that?

#### d) Run the cell below to train your model for 20 epochs. 

```python
losses = model.fit(X_train,y_train, batch_size=32, epochs=40, verbose=1)
```

## 3. Make predictions
a) Run the cell below to predict the labels on the training and test set and plot the confusion matrices. What do you observe?

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

# make predictions for test class labels
y_test_class = np.argmax(y_test, axis=1)
y_pred_test_class = np.argmax(model.predict(X_test), axis=1)

# plot confusion matrix
fig,ax = plt.subplots(figsize=(9,8.5))
conf_mat = confusion_matrix(y_test_class, y_pred_test_class, normalize="true")
_=sns.heatmap(conf_mat,xticklabels=class_names,yticklabels=class_names, annot=True, ax=ax, cmap="viridis", vmin=0.0, vmax=1.0)
_=ax.set_title("confusion matrix test set")

# make predictions for training class labels
y_train_class = np.argmax(y_train, axis=1)
y_pred_train_class = np.argmax(model.predict(X_train), axis=1)

# plot confusion matrix
fig,ax = plt.subplots(figsize=(9,8.5))
conf_mat = confusion_matrix(y_train_class, y_pred_train_class, normalize="true")
_=sns.heatmap(conf_mat,xticklabels=class_names,yticklabels=class_names, annot=True, ax=ax, cmap="viridis", vmin=0.0, vmax=1.0)
_=ax.set_title("confusion matrix training set")
```

## 2- Iterative estimation of a constant value

In this section we're going to implement an iterative algorithm to estimate the mean of a Gaussian distribution using Bayes' theorem.

a) Run the cell below to plot the target probability distribution. 

```python
from utils import plot_joint, plot_random_variables, generate_gif

# distribution's parameters
mu_target = np.array([0.6,-.5])
sigma_M = 1e-1
Sigma_target = sigma_M**2*np.eye(2)

# plot distribution
cmap_target='Reds'
color_target = 'r'
x, y = np.mgrid[-1:1:.01, -1:1:.01]
fig, ax = plt.subplots(figsize =(7,7))
plot_random_variables(ax, color_target, cmap_target, mu_target, Sigma_target, x, y, marker_size=100, marker_width=2)
```

b) Run the cell below to plot our initial guess for the mean of the target distribution.

PS.: In the plot below the contour levels show the uncertainty of our guess for the mean of the target distribution. But it is not equal to the variance of the target probability distribution!

```python
# guess' parameters
mu_0 = np.array([0,0])
Sigma_0 = 1e-1*np.eye(2)

# plot distribution
cmap='Blues'
color = 'b'
fig, ax = plt.subplots(figsize =(7,7))
plot_random_variables(ax, color, cmap, mu_0, Sigma_0, x, y, marker_size=100, marker_width=2)
```

c) Complete the function ``update()`` in the cell below. It should return the updated mean and covariance matrix of our estimate given the sampled data point.

```python

import numpy as np
from numpy.linalg import inv

def update(mu, Sigma, sigma_M, sample):
    newMu = inv(np.eye(*mu.shape)/sigma_M**2+inv(Sigma))*(1/sigma_M**2*sample+inv(Sigma)*mu)
    newSigma = inv(np.eye(*Sigma.shape)/sigma_M**2+inv(Sigma))




    return np.diagonal(newMu), newSigma  # TODO: change code here

# set random seed
np.random.seed(2)

# sample from true distribution
sample = np.random.multivariate_normal(mu_target, Sigma_target, 1)

# update estimate given sample
mu, Sigma = update(mu_0, Sigma_0, sigma_M, sample)

print("Updated mean: ")
print("Value: \n {}".format(mu))
print("Expected value: \n \
[ 0.50756747 -0.45966062]")

print("\n Updated covariance matrix: ")
print("Value: \n {}".format(Sigma))
print("Expected value: \n \
[[0.00909091 0.        ]  \n \
[0.         0.00909091]]")
```

d) Run the cell below to generate a gif with your results in the ``results/`` directory. 

```python
# delete older images if any
import os, glob
for filename in glob.glob("results/bayes_*"):
    os.remove(filename) 

# save first guess
fig, ax = plt.subplots(figsize =(7,7))
plt.scatter(0,0, color='r', label="Samples from target")
plt.scatter(0,0, color='white')
plot_random_variables(ax, color_target, cmap_target, mu_target, Sigma_target, x, y, marker_size=100, marker_width=2)
plot_random_variables(ax, color, cmap, mu_0, Sigma_0, x, y, marker_size=100, marker_width=2)
_ = ax.legend()
plt.close()
fig.savefig("results/bayes_0.jpeg", bbox_inches='tight')

# initialize variables
samples = []
iterations = 10
mu = mu_0
Sigma = Sigma_0

# set random seed
np.random.seed(2)

# generate images for gif
for i in range(1,iterations+1):

    # sample and update 
    sample = np.random.multivariate_normal(mu_target, Sigma_target, 1)
    samples.append(sample)
    mu,Sigma = update(mu, Sigma, sigma_M, sample)
    
    # plot distribution along with sample
    fig, ax = plt.subplots(figsize =(7,7))
    ax.set_xticks(np.arange(-1,1,.25))
    plt.scatter([sample[0,0] for sample in samples], [sample[0,1] for sample in samples], color='r', label="Samples from target")
    plot_random_variables(ax, color_target, cmap_target, mu_target, Sigma_target, x, y, marker_size=100, marker_width=2)
    plot_random_variables(ax, color, cmap, mu, Sigma, x, y, marker_size=100, marker_width=2)
    _ = ax.legend()
    plt.close()

    # save image
    fig.savefig("results/bayes_{}.jpeg".format(i), bbox_inches='tight')


filenames = ["results/bayes_{}.jpeg".format(i) for i in range(iterations+1)]
generate_gif(filenames, gifname="results/bayes.gif")
```

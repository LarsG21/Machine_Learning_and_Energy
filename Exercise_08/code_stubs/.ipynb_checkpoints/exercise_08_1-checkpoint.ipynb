## 1. k-Means clustering

As seen in the lecture, k-Means is a very simple yet relevant algorithm for unsupervised learning. 
In this section we are going to implement it step-by-step and we are going to use a toy dataset to visualize its functioning. 

a) Run the cell below to load and plot the toy dataset.

```python
import numpy as np
from utils import plot_dataset, generate_gif, plot_random_variables
import matplotlib.pyplot as plt

# read data from csv
X = np.genfromtxt('data/dataset.csv', delimiter=";")

# plot dataset
fig, ax = plt.subplots(figsize=(12,7))
plot_dataset(ax, X)
plt.show()
```

b) Now on ``k_means.py``, complete the code of ``find_closest_center()``.
This function should retrieve the closest center for each data point given a set of centroids. 

PS.: by closest we mean the center with shortest Eucledian distance to a data point, i.e. $\text{argmin}_i || x - \mu_i||^2_2$

```python
import k_means as kmm

# initialize centroids
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])

# find the closest centroids: verification
z = kmm.find_closest_center(X, initial_centroids)

print("Closest centroids for first 5 data points: ")
print("Value: {}".format(z[:5]))
print("Expected value: [0 2 1 0 0]")
```

c) Complete the code of ``update_centroids()``. The function estimates the new value for the centroids based on the data points that are assigned to the centroid. 

```python
# update centroids: verification 
centroids = kmm.update_centroids(X, z, k=3)

print("New estimation for centroids: ")
print("Value: \n{}".format(centroids))
print("Expected value: \n \
[[2.42830111 3.15792418] \n \
 [5.81350331 2.63365645] \n \
 [7.11938687 3.6166844 ]]")
```

d) Complete the function ``fit()``. It should perform the k-Means algorithm using the already implemented functions and should return the history of the centroids' values at each iteration. 
The cell below produces a gif with your results in the ``results/`` directory. 

```python
# run k-means
opt_centroids = kmm.fit(X, initial_centroids)

# define colors for centroids
colors = ['r', 'g', 'b']

# delete older images if any
import os, glob
for filename in glob.glob("results/kmm_*"):
    os.remove(filename) 

# generate images for gif
for i in range(len(opt_centroids)):

    # get colors
    z = kmm.find_closest_center(X, opt_centroids[i])

    # generate gif
    fig, ax = plt.subplots(figsize =(12,7))
    plot_dataset(ax, X, z, colors)
    ax.scatter(opt_centroids[i][:,0], opt_centroids[i][:,1], marker='x', color=colors, s=200, linewidths=5)
    plt.close()

    # save image
    fig.savefig("results/kmm_{}.jpeg".format(i), bbox_inches='tight')

filenames = ["results/kmm_{}.jpeg".format(i) for i in range(len(opt_centroids))]
generate_gif(filenames, gifname="results/kmm.gif")
```

## 2- EM for Gaussian mixtures

In this section we're going to implement the Expectation-Maximization (EM) algorithm for Gaussian mixtures.

a) Run the cell below to plot the initial guess of our Gaussian mixture model.
Feel free to read ``init_variables()`` on the ``em_gaussian_mixture.py`` module to understand how each variable is initialized.

```python
import em_gaussian_mixture as gmm

# initialize values
means, covs, pis = gmm.init_parameters(X, k=3, seed=6)

# plot dataset
fig, ax = plt.subplots(figsize =(12,7))
plot_dataset(ax, X)

# plot pdfs
cmaps=['Reds', 'Greens', 'Blues']
x, y = np.mgrid[-1:10:.01, 0:7:.01]
plot_random_variables(ax, colors, cmaps, means, covs, x, y)
plt.show()
```

b) Complete the function ``expectation_step()``.
This function should return the cluster assignment probabilities given the current values for mean, covariance and cluster probability.
In this step you should use the function ``multivariate_normal.pdf()`` from ``scipy.stats`` to return the probability density function evaluated at one or more points of a multivariate gaussian distribution. 

PS.: Read the documentation of ``multivariate_normal.pdf()`` to better understand its usage

```python
# expectation step
import em_gaussian_mixture as gmm
gammas = gmm.expectation_step(X, means, covs, pis)
print("Cluster assignment probabilities for first 3 data points: ")
print("Value: \n {}".format(gammas[:3,:]))
print("Expected value: \n \
[[2.25425586e-02 9.49195500e-01 2.82619416e-02] \n \
[5.40067290e-01 4.59912664e-01 2.00453418e-05] \n \
[6.96567294e-01 3.03242081e-01 1.90624761e-04]]")
```

c) Complete the function ``maximization_step()``. This function should update your model's parameters given the cluster assignment probabilities calculated in the expectation step.

```python
# maximization step
gmm.maximization_step(X, means, covs, pis, gammas)

print("Updated means: ")
print("Value: \n {}".format(means))
print("Expected value: \n \
[[6.01978994 3.0785998 ] \n \
[3.09995905 4.28017672] \n \
[2.93932139 1.0823691 ]]")

print("\n Updated covariance matrices: ")
print("Value: \n {} \n {} \n {}".format(*covs))
print("Expected value: \n \
[[ 1.32224604 -0.34734828] \n \
[-0.34734828  0.451999  ]] \n \
[[ 7.25566898 -4.22600677] \n \
[-4.22600677  3.21794125]] \n \
[[ 1.10423282 -0.5317579 ] \n \
[-0.5317579   0.92481786]]")

print("\n Updated cluster probabilities: ")
print("Value: \n {}".format(pis))
print("Expected value: \n \
[0.2178963040927625, 0.4596150175153419, 0.3224886783918956]")
```

d) Complete the function ``fit()``. It should perform the EM algorithm using the already implemented functions and should return the history of the distributions' parameters at each iteration. 
The cell below produces a gif with your results in the ``results/`` directory. 

```python
# reinitialize parameters
means, covs, pis = gmm.init_parameters(X, k=3, seed=6)

# run EM
opt_means, opt_covs, opt_pis = gmm.fit(X, means, covs, pis, max_iter=50)  # should converge in 13 iterations

# delete older images if any
for filename in glob.glob("results/gmm_*"):
    os.remove(filename) 

# generate images for gif
for i in range(len(opt_means)):

    # generate gif
    fig, ax = plt.subplots(figsize=(12,7))
    plot_dataset(ax, X)
    plot_random_variables(ax, colors, cmaps, opt_means[i], opt_covs[i], x, y)
    plt.close()

    # save image
    fig.savefig("results/gmm_{}.jpeg".format(i), bbox_inches='tight')

filenames = ["results/gmm_{}.jpeg".format(i) for i in range(len(opt_means))]
generate_gif(filenames, gifname="results/gmm.gif")
```

e) In this last item, we are going to compare our implementation of the EM algorithm for Gaussian mixtures with sklearn's one.
Give a quick look at https://scikit-learn.org/stable/modules/generated/sklearn.mixture.GaussianMixture.html and fit the Gaussian mixture model to our dataset.

```python
# import sklearn package and fit model to dataset
# TODO: add code here 
import sklearn
from sklearn.mixture import GaussianMixture

sklearn_means = np.zeros(opt_means[0].shape)  #TODO: change code here
sklearn_covs = [np.zeros(opt_covs[0][0].shape) for _ in range(len(opt_covs[0]))]  #TODO: change code here

skgmm = GaussianMixture(n_components=3, random_state=0).fit(X)

sklearn_means = skgmm.means_
sklearn_covs = skgmm.covariances_



print("Maximum difference between means: ")
print("Value: {}".format(np.max(np.max(np.abs(sklearn_means - opt_means[-1])))))
print("Expected value: 0.0011397403589556987 \n")

print("Maximum difference between covariance matrices: ")
print("Value: {}".format(max([np.max(np.abs(sklearn_covs[i] - opt_covs[-1][i])) for i in range(3)])))
print("Expected value: 0.003118935725618499")
```

As you saw, even though we didn't set the initial conditions of the sklearn's model equal to ours, we achieve pretty similar models.

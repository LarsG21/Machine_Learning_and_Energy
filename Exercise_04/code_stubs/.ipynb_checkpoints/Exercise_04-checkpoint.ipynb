## 1. A toy example

a) Run the following cell to load and plot the toy dataset. Do you think we can separate both classes with a straight line? 


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

N_TRAIN = 100  # number of training examples
NOISE_STD = 5.0  # standard deviation for Gaussian noise
SEED = 0  # seed for randomized functions

def toy_fcn(x, sigma=NOISE_STD, seed=0):
    return np.multiply(x[:,1] > (-1/0.201)*(0.206*x[:,0] -25.161) + np.random.RandomState(seed).normal(0, sigma, x.shape[0]), 1)

# Generate data
X = np.random.RandomState(SEED).uniform(low=30,high=100,size=[N_TRAIN,2])
y = toy_fcn(X, NOISE_STD, seed=SEED)

# Plot data
plt.figure(figsize=(15,7))
plt.scatter(*X[y==0].T, color="red", label="y=0")
plt.scatter(*X[y==1].T, color="blue", label="y=1", marker="P")
_ = plt.legend()
```

c) In the classification task, we generally use cross-entropy as the loss function and it is given by: 
$$J(x) = - \sum_{x \in \mathcal{X}}  p(x) \log q(x).$$

Cross-entropy was originally designed to be used in information theory to quantify the message overhead needed to code an event drawn from a distribution $p(x)$ when the coding scheme was optimized for a different distribution (with same support $\mathcal{X}$) $q(x)$.
In the case where $p(x)$ is either 1 or 0, we can simplify the above expression as 
$$J(x) = -p(x)\log q(x) - (1-p(x))\log (1-q(x)).$$

In the cell below complete ``cross_entropy()``, which receives values for the true and estimated distributions and calculates their cross-entropy.
What can you observe in the plot below? Can you say now why we use cross-entropy in the classifcation task?

```python
def cross_entropy(p,q): 
    return -p*np.log(q)-(1-p)*np.log(1-q)  # <-- change this

# Define test points
q = np.arange(0,1,step=0.01)

# Evaluate cross entropy for p=1 and p=0
cost_pos = cross_entropy(1,q)
cost_neg = cross_entropy(0,q)

# Plot graph
plt.figure(figsize=(15,7))
plt.plot(q, cost_pos, color="blue", label="p=1")
plt.plot(q, cost_neg, color="red", label="p=0")
_ = plt.legend()
```

d) Complete ``cost_function()`` in ``logistic_regression``.
Remember that our estimated function is given by $\sigma (\mathbf{w}^T\mathbf{x})$.
Also, average the cross-entropy over the $N$ training points to obtain our loss estimator.

```python
# Check cost
w_test = np.array([-24, 0.2, 0.2])
cost, _ = loreg.cost_function(w_test, X, y)

print(f"Cost at test weights: {cost}")
print(f"Expected cost (approx): 0.139")
```

e) To minimize our loss function we will use the gradient descent implementation of ``scipy.optimize``.
The gradient for our loss function is given by
$\frac{\partial J(w)}{\partial w_j}= \frac{1}{n} \sum_{i=1}^{n} (\sigma(\mathbf{w}^T \mathbf{x}_i)-y_{i})x_{i,j}$.
Complete the code in ``cost_function()`` so that it also returns the gradient of our cost function. 
Test your implementation by running the cell below. 

```python
# Check gradient
w_test = np.array([-24, 0.2, 0.2])
[cost, grad] = loreg.cost_function(w_test, X, y)

print(f"Gradient at test weights: {grad}")
print(f"Expected gradients (approx): [0.017 1.09 0.658]")
```

f) Run the cell below to obtain the optimal weigths ``w``.

```python
# Optimize w
w,loss = loreg.fit(X,y)

# Check weights found by optimization
print(f"Cost at w: {loss}")
print("Expected cost (approx): 0.114")
print(f"w: {w}")
print("Expected w (approx): [-44.379 0.358 0.373]")
```

g) Run the cell below and visualize the optimal decision boundary determined by the weights obtained in the previous cell.

```python
# Plot Boundary
fig=plt.figure(figsize=(15,7))
plt.scatter(*X[y==0].T, color="red", label="y=0")
plt.scatter(*X[y==1].T, color="blue", label="y=1", marker="P")
plot_x = np.array([np.min(X[:,0])-2, np.max(X[:,0])+2])
plot_y = (-1/w[2])*(w[1]*plot_x + w[0])
plt.plot(plot_x, plot_y) 
_ = plt.legend()
```

h) Complete the function ``predict()``, which takes a matrix $\mathbf{X}$, the weights $\mathbf{w}$ and the threshold parameter $t$ as inputs. 
This function computes the class probabilities and classifies the samples of $\mathbf{X}$ depending on the chosen threshold.
Run the cell below to verify the accuracy of the model.

```python
# Compute accuracy on training set
y_hat = loreg.predict(w,X,t=.5) # prediction for class

accr=np.mean(y_hat == y) * 100 # compute accuracy

# Check accuray
print(f"Train Accuracy: {accr}")
print("Expected accuracy for t=0.5(approx): 94.0")
```

i) Sometimes when evaluating a classification model we need more information than only the model accuracy.
One way of doing that is using a confusion matrix, in which rows show the predicted class and columns the true class of the evaluated samples.
E.g., in the case of binary classification, the value in the first row and first column shows the number of samples correctly predicted in class 0.
A model with 100\% accuracy would have values different than zero only in the matrix's main diagonal.
Complete the code of ``get_confusion_matrix()`` so it returns the confusion matrix for given predictions ``y_hat`` and true values ``y`` as a ``numpy.array``.

```python
import seaborn as sns

# Build confusion matrix
mx = loreg.get_confusion_matrix(y_hat,y)

# Build dataframe and plot
df_cm = pd.DataFrame(mx, range(2), range(2))
sns.set(font_scale=1.4)
fig, ax = plt.subplots()
sns.heatmap(df_cm, annot=True, cmap=sns.cm.rocket_r, cbar=False)
plt.xlabel("Actual class")
plt.ylabel("Predicted class")
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
fig.show()
```

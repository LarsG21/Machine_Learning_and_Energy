In this notebook we will use logistic regression to predict the type of fault in a power system. 
Here we divide faults into two classes: transient and non-transient. 
Transient faults are assumed to vanish after disconnecting the system for a short period of time and then connecting it back again.
This type of faults can happen for several reasons, like lightining strikes or momentary contact with external objects. 
Non-transient faults, however, are still present after reconnection and therefore need to be actively cleared by the operator.

## 1. The dataset

The dataset we will use contains snapshots of the power system status when a fault happened. 
Each snapshot has information about the weather condition (0=clear, 5=storm), whether the protection system emitted a trip signal, the current (A) and the voltage (kV) at the line, and the frequency (Hz).
Transient faults are marked with 1, while non-transient ones with 0.

a) Read the dataset from ``dataset.csv`` and print the first 10 rows.


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

# Loading dataset
dataset = pd.read_csv("dataset.csv",delimiter=';')
dataset.head(10)

# Print first 10 rows of dataset
# Add your code here
```

b) Run the cell below. What can you say about the data? 

```python
import seaborn as sns

sns_plot = sns.pairplot(dataset, height=2, corner=True)
```

c) Split the data into input array ``X`` and target class ``y`` by completing the code in the cell below. We've already added the code to split the arrays into train and test set. <br>
#### OBS.: As the number of samples in the dataset is quite small, we utilized a ratio of 70/30 of train and test data so we have enough samples to evaluate our model.

```python
# Split dataset
X = dataset.drop(['is_transient'], axis=1).values  #<--
y = dataset['is_transient'].values      #<--

# Define constants
N_SAMPLES = X.shape[0]
SAMPLES_TRAIN = round(.7*N_SAMPLES)
SAMPLES_TEST = N_SAMPLES - SAMPLES_TRAIN

# Split data points
X_train = X[:SAMPLES_TRAIN]
y_train = y[:SAMPLES_TRAIN]
X_test = X[-SAMPLES_TEST:]
y_test = y[-SAMPLES_TEST:]
```

## 2. Training logistic regression model

Here we will use our logistic regression code in ``logistic_regression`` to classify the faults in the power system.

a) In the cell below, fit our model using the training set. We have already normalized the features for you :)

```python
import my_utils 
from scipy.optimize import minimize
import logistic_regression as loreg

# Normalize features
mu_X = np.mean(X_train, axis=0)
sigma_X = np.std(X_train, axis=0)
X_train = my_utils.normalize_features(X_train, mu_X, sigma_X)

# Optimize w
w,loss = loreg.fit(X_train,y_train)

# Check weights found by optimization
print(f"Cost at w: {loss}")
print("Expected cost (approx): 0.393")
print(f"w: {w}")
print("Expected w (approx): [-1.121 0.139 1.432 0.979 -1.00 0.870]")
```

## 3. Testing model

a) Run the cell below to verify the accuracy of our model. 

```python
# Normalize test set
X_test = my_utils.normalize_features(X_test, mu_X, sigma_X)

# Compute accuracy on training set
y_hat = loreg.predict(w,X_test,t=0.5) # prediction for class

accr = np.mean(y_hat == y_test) * 100 # compute accuracy

# Check accuracy
print(f"Train Accuracy: {accr}")
print("Expected accuracy for t=0.5(approx): 86.96")

# Build confusion matrix
mx = loreg.get_confusion_matrix(y_hat,y_test)

# Build dataframe and plot
df_cm = pd.DataFrame(mx, range(2), range(2))
sns.set(font_scale=1.4)
fig, ax = plt.subplots()
sns.heatmap(df_cm, annot=True, cmap=sns.cm.rocket_r, cbar=False)
plt.xlabel("Actual class")
plt.ylabel("Predicted class")
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
fig.show()
```

## 4. Using sklearn

Now we want to compare our model with the logistic regression implementation of ``sklearn``.

a) Use ``LogisticRegression()`` with no penalty (have a look at the function docu) to set the model. ``X_train_sklearn`` and ``y_train`` are the input and target data, respectively. 
#### HINT: we did something similar in the last exercise ;) 

```python
from sklearn.linear_model import LogisticRegression 

X_train_sklearn = np.concatenate((np.ones([X_train.shape[0],1]), X_train), axis=1)
X_test_sklearn = np.concatenate((np.ones([X_test.shape[0],1]), X_test), axis=1)

# Fit model on train set
mdl_lr = LogisticRegression(penalty = 'none').fit(X_train_sklearn, y_train)  # <-- change this

# Get optimal weights
w_lr = mdl_lr.coef_ # <-- change this

print(f"w: {w_lr}")
print("Expected w (approx): [-0.560 0.139 1.432 0.979 -1.003 0.870]")
```

b) Make predictions on the test data ``X_test_sklearn`` and plot its confusion matrix.

```python
# Make predictions on test set
y_pred_lr = mdl_lr.predict(X_test_sklearn)

accr = np.mean(y_pred_lr == y_test) * 100 # compute accuracy

# Check accuracy
print(f"Train Accuracy: {accr}")
print("Expected accuracy for t=0.5(approx): 86.96")

# Build confusion matrix
mx = loreg.get_confusion_matrix(y_pred_lr,y_test)

# Build dataframe and plot
df_cm = pd.DataFrame(mx, range(2), range(2))
sns.set(font_scale=1.4)
fig, ax = plt.subplots()
sns.heatmap(df_cm, annot=True, cmap=sns.cm.rocket_r, cbar=False)
plt.xlabel("Actual class")
plt.ylabel("Predicted class")
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
fig.show()
```

## 2. The data set
In the folder ``/data`` there are 6 panda dataframes stored in the .pkl format:
- ``data_train_1.pkl``, ``data_train_2.pkl``,``data_train_3.pkl``: These data frames form the training data. They contain values for u10, v10m u100, v100 and power_generation from 2012/01/01 to 2013/08/31 for three sites.


- ``data_test_1.pkl``, ``data_test_2.pkl``, ``data_test_3.pkl``: These data frames form the test data. They contain only values for u10, v10m u100, v100. This data ranges from 2013/09/01 to 2013/12/30. We will reveal the missing target values for this data after the competition.

Let's have a look at the data:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data_train_1 = pd.read_pickle("data/data_train_1.pkl")
data_train_2 = pd.read_pickle("data/data_train_2.pkl")
data_train_3 = pd.read_pickle("data/data_train_3.pkl")
data_train_1.head()
```

Each data frame has the following columns:
- datetime: timestamp with date and hour information
- u10: the forecasted meridional wind component (i.e. wind from East->West) at 10 meters above the ground
- u100: the forecasted meridional wind component (i.e. wind from East->West) at 100 meters above the ground
- v10: the forecasted zonal wind component (i.e. wind from South->North) at 10 meters above the ground
- v100: the forecasted zonal wind component (i.e. wind from South->North) at 100 meters above the ground
- power_generation: the normalized hourly energy output of the wind turbine (only available for the training data)



Let's have a look at the generation:

```python
fig, axs=plt.subplots(3,1, figsize=(16,8))
data_train = [data_train_1,data_train_2,data_train_3]
for i in [0,1,2]:
    axs[i].plot(data_train[i]["power_generation"]*10, label="power generation")
    axs[i].plot(data_train[i]["u10"], label="u10")
    axs[i].plot(data_train[i]["v10"], label="v10")
    axs[i].set_title(f"Site {i+1}")
    axs[i].legend()
fig.tight_layout()
```

Now let's take a look at the test data:

```python
data_test_1 = pd.read_pickle("data/data_test_1.pkl")
data_test_2 = pd.read_pickle("data/data_test_2.pkl")
data_test_3 = pd.read_pickle("data/data_test_3.pkl")
data_test_1.head()
```

As you can see, the test data is missing the last column with the generated power.

**The task for this project is to forecast the hourly normalized wind power genration for three different sites for the test data.**

The evaluation is based on the RMSE. 
Let $y_{i,t}$ be the actual value for the wind power output for site $i$ at time $t$ and let $\hat{y}_{i,t}$ the predicted values, then the RMSE is given by
$$RMSE=\sqrt{\frac{1}{NT}\sum_{i=1}^N\sum_{t=1}^T (y_{i,t}-\hat{y}_{i,t})^2}.$$

## 2. Model Building
Building on what we have learned from the exploratory data analysis, we prepared the following code to create data frames for the design matrix $\mathbf{\Phi}$ and the vector $\mathbf{y}$.
#### a) Given the code below, write down the formula of the corresponding model in the form $\hat{y}_t = w_0 + w_1f_1(x_t) + w_2f_2(x_t) + ...$ (of course you don't have to write down all 548 parts explicitly). Do you think the choice of features is reasonable?

y = ((Phi^T*Phi+Lamda*I)*w)/Phi^T    --> Phi = matrix (548 x deg)  --> deg = gewnschete Dimension der Nherung

                                           --> Ein Datenpunkt hat 548 Dimensionen

```python
# create squared temperature feature
data["temp2"] = data["temp"].pow(2)

# create lagged load features
LAGGED_DAYS = 7
lags_load = np.arange(HOUR,24*LAGGED_DAYS+HOUR,1)
for l in lags_load:
    data["load_t-"+str(l)] = data["load"].shift(l)

# create lagged temperature features
lags_temp = np.arange(1,24*LAGGED_DAYS+HOUR,1)
for l in lags_temp:
    data["temp_t-"+str(l)] = data["temp"].shift(l)
    data["temp2_t-"+str(l)] = data["temp2"].shift(l)

#cut off NA values from lagged features
data = data.iloc[LAGGED_DAYS*24:,:]

# add one-hot encoded features for month and weekday
MoY = pd.get_dummies(data[["month"]], prefix=None, drop_first=True)
DoW = pd.get_dummies(data[["weekday"]] , prefix=None, drop_first=True)
data = pd.concat((data, MoY, DoW), axis=1)

# create Phi(our input features) and y (ourtarget) from data for hour 13
data = data[data.hour==HOUR]
Phi_df = data.drop(["load", "hour", "weekday", "month"], axis=1)
y_df = data["load"]
Phi_df.head(5)
```

#### b) Complete the code in the cell below.
Use scikit learns ``LinearRegression``, ``Lasso``, and ``Ridge`` class to train an unregularized, an L1 regularized, and an L2 regularized linear regression model using the data in ``Phi_train`` and ``y_train``. Then make predictions for the test data in ``Phi_test``.
Compare the obtained model weights and the predictions for the test data.


```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

def compute_rmse(true, pred):
    return np.sqrt(np.mean((true-pred)**2))

# split in train and test data
Phi_train = Phi_df.loc['2006-01-01':'2007-12-31'].values
Phi_test = Phi_df.loc['2008-01-01':'2008-12-31'].values
y_train = y_df['2006-01-01':'2007-12-31'].values
y_test = y_df['2008-01-01':'2008-12-31'].values


# Linear Regression model
mdl_lr = LinearRegression(normalize=True).fit(Phi_train, y_train)
y_pred_lr = mdl_lr.predict(Phi_test)
w_lr = mdl_lr.coef_

# ------------ add code here -----------------------------------

# Lasso model
lambda_lasso = 0.1
mdl_lasso = Lasso(alpha=lambda_lasso, normalize=True).fit(Phi_train, y_train) # <-- change this
y_pred_lasso = mdl_lasso.predict(Phi_test) # <-- change this
w_lasso = mdl_lasso.coef_ # <-- change this

# Ridge model
lambda_ridge = 0.001
mdl_ridge = Ridge(alpha=lambda_ridge, normalize=True).fit(Phi_train, y_train) # <-- change this
y_pred_ridge = mdl_ridge.predict(Phi_test) # <-- change this
w_ridge = mdl_ridge.coef_ # <-- change this

# ---------------------------------------------------------------


# plot predictions for the first month of 2018
fig,ax = plt.subplots(figsize=(16,8))
pred_df = pd.DataFrame(data=np.column_stack((y_test, y_pred_lr, y_pred_ridge, y_pred_lasso)), index=y_df['2008-01-01':'2008-12-31'].index, columns=["true", "lr", f"ridge ({lambda_ridge})", f"lasso ({lambda_lasso})"])
sns.lineplot(data=pred_df['2008-01-01':'2008-01-31'], linewidth=2.0, ax=ax)
ax.set_ylabel("load")
fig.tight_layout()
#plt.savefig("lr_lasso_ridge.jpeg")
```

Both regularized models seem to work quite well. So let's try to find the optimal degree of regularization.
#### c) Complete the code in the cell below.
The code tests different combinations for the ratio between L1 and L2 penalty (``l1_ratios``) and strength of the penalty term (``lambdas``) for an elastic net model (i.e. a model that combines L1 and L2 regularization) and saves the models and the corresponding test set RMSEs in the respective dataframes.

```python
from sklearn.linear_model import ElasticNet
import sklearn
# possible hyper parameters
l1_ratios = np.linspace(0.0, 1.0, 11, endpoint=True)
lambdas = np.power(10.0, - np.linspace(0,6,13, endpoint=True))

# initialize empty data frames to store models and errors
models = pd.DataFrame(index=l1_ratios, columns=lambdas)
rmse = pd.DataFrame(index=l1_ratios, columns=lambdas, dtype=np.float)

# ------------ add code here -----------------------------------
# loop over hyper parameter sets
for l1_ratio in l1_ratios:
    for lam in lambdas:
        model = ElasticNet(alpha=lam ,normalize = True, l1_ratio=l1_ratio).fit(Phi_train,y_train)
        models.loc[l1_ratio, lam] = model
        rmse.loc[l1_ratio, lam] = sklearn.metrics.mean_squared_error(y_true=y_test, y_pred=model.predict(Phi_test), squared=False)
# ---------------------------------------------------------------

# plot results
fig,ax = plt.subplots(figsize=(9,8))
sns.heatmap(rmse.values, xticklabels=lambdas.round(7), yticklabels=l1_ratios.round(4), ax=ax, cmap="jet")
plt.ylabel("l1 ratio")
plt.xlabel("lambda")
_=ax.set_title("RMSE")
fig.tight_layout()
#plt.savefig("en_results.jpeg")
```

#### d) Run the cell below to visualize the sparsity pattern for the weights of the lasso models. Connect the plot to the formula of your model. Do you find the result of the feature selction surprising?

```python
# get list of weights from pure lasso models
W = []
for lam in lambdas[0:8]:
    W.append(models.loc[l1_ratios[-1],lam].coef_)

# plot
fig,ax = plt.subplots(figsize=(20,3))
sns.heatmap( np.vstack(W)!=0, ax=ax, cmap="binary", linecolor="gray", linewidths=0.005, cbar=False, yticklabels=lambdas.round(6)[0:8])
ax.set_ylabel("lambda")
ax.set_xlabel("weights")
ax.set_title("sparsity pattern from Lasso regression (non-zero weights shown in black)")
fig.tight_layout()
#plt.savefig("sparsity.jpeg")
```

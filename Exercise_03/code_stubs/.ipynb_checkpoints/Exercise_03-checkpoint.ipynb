## 1. A simple toy problem
Let's get started by looking at a simple, one dimensional toy problem.
#### a) Run the cell below to create our example and plot it.

```python
import numpy as np
import matplotlib.pyplot as plt
import LinearRegression as lr

# set some parameters
N_TRAIN = 15    # number of training examples
NOISE_STD = 1.0 # standard deviation for Gaussian noise

# toy function
def toy_fcn(x, sigma=NOISE_STD, seed=42):
    return 3 + 0.7*x + 0.2*x**2 + 2.0*np.sin(1.5*x) + np.random.RandomState(seed).normal(0, sigma, x.shape)

# draw a lot of samples from the true function
X_true = np.expand_dims(np.linspace(-5.0,5.0,500),1)
y_true = toy_fcn(X_true, sigma=0)

# but we only obtain some noisy observations from the function (the training data):
X_train = np.expand_dims(np.linspace(-5.0, 5.0, N_TRAIN),1) #+ np.random.RandomState(123).normal(0,0.2,(N_TRAIN,1))
y_train = toy_fcn(X_train)

# plot true function and training data
plt.figure(figsize=(15,7))
plt.plot(X_true, y_true, label="true function", linestyle="--", color="gray")
plt.scatter(X_train, y_train, label="training data", color="blue")
plt.ylim(np.min(y_true)-2, np.max(y_true)+2)
_=plt.legend()
```

In the plot you can see the function $f(x)= 3 + 0.7x + 0.2x^2 + 2sin(1.5x)+ \epsilon$ (where $\epsilon$ is normally distributed noise) as grey dashed line.

From this function, we obtained a set of 15 noisy observations (blue dots), i.e. pairs of x- and y-values.
We call these observations our *training data* or *training set*.

Now, one could ask:

*"Given our training data of x-y-pairs, what is a good prediction $\hat{y}$ for $x=2.1$?"*

In linear regression the answer is: We define a linear function $f(x)=\sum_i w_if_i(x)$ that approximates the underlying function well enough and evaluate our function at point $x=2.1$.

In matrix form we can write this linear model as $\mathbf{y} = \mathbf{\Phi}\mathbf{w}$ with $\mathbf{y} \in \mathbb{R}^{N \times 1}, \mathbf{\Phi} \in \mathbb{R}^{N \times M}, \mathbf{w} \in \mathbb{R}^{M \times 1}$.

## 2. Model training and making predictions
Training our linear regression model corresponds to finding the optimal values for $\mathbf{w}$. This is quite simple as the normal equation provides an analytic solution to the optimization problem.
#### a) Complete the function ``fit()`` in the module ``LinearRegression``.
The function computes the weight vector $\mathbf{w}$ for a given design matrix $\mathbf{\Phi}_{train}$ and corresponding target values $\mathbf{y}_{train}$ using the normal equation.

NOTE: We usually do not want to penalize the value of the constant (i.e. the first entry in $\mathbf{w}$). To this end, use the modified identity matrix ``eye``.

Run the cell below to check your implementation.

```python
weights = lr.fit(np.concatenate((np.ones((X_train.shape[0],1)), X_train), axis=1), y_train, lam=1.0)
print(f"weights:  {weights[:,0]}\nexpected: [4.91511043 0.55207794]")
```

#### b) Complete the function ``predict()`` in the module ``LinearRegression``.
It computes the prediction $\mathbf{y}_{test}$ from the estimated weights $\mathbf{w}$ and the design matrix $\Phi_{test}$.

Run the cell below to check your implementation.

```python
pred = lr.predict(np.array([[1, 2.1]]), weights)
print(f"prediction:{pred[0,0]: .4f}\nexpected:   6.0745")
```

## 4. Regularization

Obviously $\lambda$ has quite an effect on our model. Let's take a closer look on this effect:
#### a) Run the cell below to fit the polynomial model with different values for $\lambda$. Explain the effect of $\lambda$ on the training and test error shown in the plot. What seems to be a reasonable value for $\lambda$?


```python
def compute_mse(true, pred):
    return np.mean((true-pred)**2)

# generate test data
X_test = np.expand_dims(np.linspace(-5.0, 5.0, 500),1)
y_test = toy_fcn(X_test)

# loop over different lambdas
lambdas = np.power(10.0,-np.arange(-1,10,1))[::-1]
err_train = []
err_test = []
for lam in lambdas:
    err_test.append(compute_mse(y_test, models.model_3(X_test, X_train, y_train, deg=DEG, lam=lam)))
    err_train.append(compute_mse(y_train, models.model_3(X_train, X_train, y_train, deg=DEG, lam=lam)))

# plot training and test error
plt.figure(figsize=(13,6))
plt.plot(lambdas, err_test, label="test")
plt.plot(lambdas, err_train, label="train")
plt.xscale("log")
plt.xticks(lambdas)
plt.xlabel("lambda")
plt.ylabel("MSE")
plt.legend()
plt.tight_layout()
#Overfitted at the Left !                     Quite good for Lamda = 10^-3                     Underfitted on the Right
#Her we would see just a wiggly Cure around the true Function                                               Here we would see just a linear approximation
#Called Bias-Variance Balanve
#On the right high Bias on the Right high Variance
```

Finally, let's explore the effect of gathering more training data. Therefore, we will compute the MSE for each of our models for different training set sizes. To account for the effect of randomness in generating the training data, we will repeat the experiment 100 times and average the resulting error.

#### b) Run the code below to run the experiment. Explain the shape of each of the curves in the plot.

```python

# generate more training data
X_train_2 = np.expand_dims(np.linspace(-5.0, 5.0, 100),1)
LAMBDA = 0.01
N = np.arange(5,X_train_2.shape[0]+1,5)
M = 100

# loop over training set sizes of different size M times
MSEs = {"linear": np.zeros((len(N),M)), "correct": np.zeros((len(N),M)), "poly": np.zeros((len(N),M)), "poly_reg": np.zeros((len(N),M))}
for j in range(0,M):
    np.random.shuffle(X_train_2)
    y_train_2 = toy_fcn(X_train_2)
    for i,n in enumerate(N):
        MSEs["linear"][i,j] = compute_mse(y_test, models.model_1(X_test, X_train_2[0:n,:], y_train_2[0:n,:]))
        MSEs["correct"][i,j] = compute_mse(y_test, models.model_2(X_test, X_train_2[0:n,:], y_train_2[0:n,:]))
        MSEs["poly"][i,j] = compute_mse(y_test, models.model_3(X_test, X_train_2[0:n,:], y_train_2[0:n,:], deg=DEG))
        MSEs["poly_reg"][i,j] = compute_mse(y_test, models.model_3(X_test, X_train_2[0:n,:], y_train_2[0:n,:], deg=DEG,lam=LAMBDA))

# plot MSE against training set size
fig, ax = plt.subplots(figsize=(12,7))
for key in MSEs:
    ax.plot(N, np.median(MSEs[key],axis=1), label=key)
    ax.fill_between(N, np.quantile(MSEs[key],axis=1, q=0.2), np.quantile(MSEs[key],axis=1, q=0.8), alpha=0.2)
ax.hlines(NOISE_STD, N[0], N[-1], linestyle="--", color="grey", label="irreducible error")
ax.legend()
ax.set_xticks(ticks=N)
ax.set_ylim((0.0, 8.0))
ax.set_xlabel("training set size")
ax.set_ylabel("median MSE")
ax.set_title(f"Median test set MSE from {M} runs")
fig.tight_layout()
```
